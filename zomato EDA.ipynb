```python
#show number of missing values in a column
df.isnull().sum()
```

```python
#checking missing values using list comprehension
[features for features in df.columns if df[features].isnull().sum()>1]
```

```python
matplotlib.rcParams['figure.figsize']= (12,6) #show bigger size image
#checking missing values using heat map
sns.heatmap(df.isnull(), yticklabels= False, cbar= False, cmap= 'viridis')
```

```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
df.columns
```

```python
#Combining two dataframes by columns
final_df = pd.merge(df,df_country, on= 'Country Code', how = 'left')
final_df
```

```python
#to check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
#we are trying to find ot with respect to country column how many records are there
final_df.Country.value_counts()
```

through above data we can conclude that people in india use zomato app more 

```python
#to show all the country
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
# Pie chart - Top 3 countries that uses zomato
plt.pie(country_val[:3], labels = country_names[:3], autopct='%1.2f%%')
```

Observation : Zomato maximum transction are from india
After that USA and then United Kingdom

```python
final_df.columns
```

```python
#Now we want to check about the rating data
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
#renaming the last column to Rating Count
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

### OBSERVATION
maximum number of peole have not rated zomato

1. When rating is between 4.5 to 4.9 ------> Excellent
2. When rating is between 4.0 to 4.4 ------> very good
3. When rating is between 3.5 to 3.9 ------> Good
4. When rating is between 2.5 to 3.4 ------> Average
5. When rating is between 1.8 to 2.4 ------> Poor

```python
ratings.head()
```

```python
#now we plot a bar graph between aggregate rating and rating count
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', data = ratings)
matplotlib.rcParams['figure.figsize']= (12,6) #show bigger size image
```

```python
#in label box color and color name are not matching
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', hue = 'Rating color', data = ratings)
```

```python
ax = plt.axes() 
# Set Inner Color
ax.set_facecolor('black') 
#now we will map the colors
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', hue = 'Rating color', data = ratings, palette= ['white','red','orange','yellow','green','green'])
```

observation:


    1. Not rated count is very high
    2. Maximum number of ratings are between 2.5 to 3.4 

```python
ax = plt.axes() 
# Set Inner Color
ax.set_facecolor('black') 
# count plot
sns.countplot(x = 'Rating color', data =ratings, palette= ['white','red','orange','yellow','green','green'])
```
